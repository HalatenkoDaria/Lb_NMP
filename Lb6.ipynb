Завдання: Побудувати інтерполяційний багаточлен Лагранжа Ln(x) для функції f(x), що задана таблицею, та з точністю до 0.001 обчислити наближені значення функції у заданих точках. Побудувати графік інтерполяційної функції y=Ln(x) за наявним набором точок. 
\
№1 Задана таблиця значень функції:
\
xi    -2 -1 0 1\
f(xi) -7  4 1 2\
Задані точки:\
xi    -3 -1,5 0,5 1,5\
f(xi)  ?   ?   ?   ?

```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.interpolate import lagrange
x = [-2, -1, 0, 1]
y = [-7, 4, 1, 2]
points = [-3, -1.5, 0.5, 1.5]

def lagrange_interpolation(x, y, xi):
    n = len(x)
    result = 0.0
    for i in range(n):
        term = y[i]
        for j in range(n):
            if j != i:
                term *= (xi - x[j]) / (x[i] - x[j])
        result += term
    return result

# Обчислення значень функції в заданих точках
approximated_values = [lagrange_interpolation(x, y, xi) for xi in points]

# Виведення результатів
print("Значення функції у заданих точках:")
for xi, fi in zip(points, approximated_values):
    print(f"f({xi}) = {fi:.3f}")

# Побудова графіка
x_range = np.linspace(min(x), max(x), 100)
y_range = [lagrange_interpolation(x, y, xi) for xi in x_range]

plt.figure(figsize=(8, 6))
plt.plot(x_range, y_range, label='Інтерполяція', color='aqua')
plt.scatter(x, y, label='Точки даних', color='red')
plt.scatter(points, approximated_values, label='Оцінені точки', color='orange')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Графік інтерполяції 1')
plt.legend()
plt.grid(True)
plt.show()
```
